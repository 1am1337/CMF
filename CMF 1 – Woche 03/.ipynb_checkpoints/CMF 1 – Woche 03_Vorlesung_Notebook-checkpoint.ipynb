### Computergestützte Fugenanalyse

Ziel: Bestimme, wie oft das Fugenthema in der ersten Sopranstimme des Chor-Fuge *Omnes Generationes* von J. S. Bach vorkommt.

```python
import music21 as m21
import matplotlib.pyplot as plt
```

```python
# Lade die MIDI-Datei 'bach_omnes-generationes.mxl' in einen m21-Score-Stream
omnes = m21.converter.parse('bach_omnes-generationes.mxl')
```

```python
# Schaue dir die Namen der einzelnen Stimmen an
for part in omnes.parts:
    print(part.partName)
```

```python
# Lade den Part-Stream, der der ersten Sopran-Stimme sntspricht
soprano_part = omnes.parts[0]
```

```python
# Erstelle eine "flatted"-Version dieses Part-Stream
# (wichtig, um auf die Offsets aller Noten mit Bezug auf den Anfang des Stückes zugreifen zu können)
flattened_soprano_part = soprano_part.flatten()
```

```python
# Extrahiere alle in der Sopran-Stimme enthaltenen Noten
soprano_notes = flattened_soprano_part.getElementsByClass(m21.note.Note)
```

```python
# Stelle die Sopran-Stimme als eine Abfolge von Intervallen dar
soprano_intervals = []
previous_midinumber = soprano_notes[0].pitch.midi

for note in soprano_notes[1:]:
    current_midinumber = note.pitch.midi
    current_interval = current_midinumber - previous_midinumber
    soprano_intervals.append(current_interval)
    previous_midinumber = current_midinumber
```

```python
print(soprano_intervals)
```

Ziel des nächsten Schrittes ist es, die im Fugenthema enthaltenen Intervalle mit den Intervallen in der Sopran-Stimme abzugleichen. \
Definiere eine Function *pattern_finder*, die zählt wie oft ein bestimmtes Intervall-Pattern (gegeben in Form einer Liste) innerhalb einer Liste von Intervallen vorkommt. \
Beispiel: Das Intervall-Pattern [1,2,3] kommt 3-Mal in der Intervall-Liste [0,1,2,3,4,1,2,3,1,2,1,2,3] vor.

```python
def pattern_finder(interval_list, interval_pattern):

    # Wie oft kommt das Pattern interval_pattern in der Liste interval_list vor:
    number_of_found_patterns = 0
    # Zu welchen Indezes kommt das Pattern interval_pattern in der Liste interval_list vor:
    list_with_pattern_indices = []
    
    len_interval_list = len(interval_list)
    len_interval_pattern = len(interval_pattern)

    for i in range(len_interval_list):

        if interval_list[i : i+len_interval_pattern] == interval_pattern:   
        
            number_of_found_patterns += 1
            list_with_pattern_indices.append(i)
            
    return number_of_found_patterns, list_with_pattern_indices
```

In unserem Fall genügt es, das Fugenthema auf die Intervall-Patterns [0, 0, 0, 0, -1] und [0, 0, 0, 0, -2] zu reduzieren. \
Wende die Funktion *pattern_finder* an, um zu zählen, wie oft das Fugenthema in der ersten Sopranstimme des Chors *Omnes Generationes* vorkommt.

```python
# determine the total number of occurences of the fugue subject in the soprano part
n_found_subjects = pattern_finder(soprano_intervals, [0, 0, 0, 0, -1])[0] + pattern_finder(soprano_intervals, [0, 0, 0, 0,- 2])[0]
```

```python
print("Das Fugenthema kommt %i-mal in der ersten Sopranstimme vor." % n_found_subjects)
```

```python
subject_indices = pattern_finder(soprano_intervals, [0, 0, 0, 0, -1])[1] + pattern_finder(soprano_intervals, [0, 0, 0, 0, -2])[1]
```

```python
subject_offsets = []

for index in subject_indices:
    subject_offsets.append(soprano_notes[index].offset)
```

```python
subject_measures = [offset/4 + 1 for offset in subject_offsets]
print(subject_measures)
```

```python
x = subject_measures
y = ["Soprano 1"]*len(x)

plt.style.use('dark_background')

fig, ax = plt.subplots(figsize=(10, 2.5))

ax.set_frame_on(False)

ax.scatter(x, y, c='plum', marker='|')
ax.set_xlabel('Measures', fontsize=10)

plt.show()
```
