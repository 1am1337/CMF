### Methode 1: Note-Extraction mithilfe von *.flatten()*

Die Methode *.flatten()* ist eine wichtige Methode, um auf alle in einem Stream enthaltenen musikalischen Elemente und Parameter zugreifen zu können.\
*.flatten()* erlaubt es, einen music21-*Score* "abzuflachen", in dem es die verschachtelte Strukturen von Streams und Substreams in einen einzigen Stream umwandelt.\
Ein Vorteil: die Offsets aller Objekte werden in Bezug auf den Anfang des Stücks angegeben.  

4) Erstelle eine *flattened*-Version der Soprano-Stimme.

```python
soprano_part_flattened = soprano_part.flatten()
```

```python
#soprano_part_flattened.show('text')
```

5) Nutze die *flattened*-Sopranstimme, um alle *m21.note.Note*-Objekte aus der Sopran-Stimme zu extrahieren.\
Packe diese in eine Liste (z. B. mithilfe der Python-*list*-Funktion)

```python
soprano_notes = list(soprano_part_flattened.getElementsByClass(m21.note.Note))
```

```python
print(soprano_notes)
```

6) Kurze Zwischen-Übung:\
Greife auf die 50. Note der Sopran-Stimme zu.\
Bestimme den Namen dieser Note (z. B. mithilfe der *Note*-Methode *name*).\
Bestimme den Offset (in Viertelnoten ausgedrückt) dieser Note in Bezug auf den Anfang der Fuge (mithilfe der *Note*-Methode *offset*).

```python
fiftieth_note = soprano_notes[49]

fiftieth_note_name = fiftieth_note.name
print(fiftieth_note_name)

fiftieth_note_offset = fiftieth_note.offset
print(fiftieth_note_offset)
```

7. Erstelle eine Liste, die alle Notennamen der Sopran-Stimme enthält.

```python
soprano_note_names = []

for note in soprano_notes:
    note_name = note.name
    soprano_note_names.append(note_name)
```

```python
print(soprano_note_names)
```

8) Iteriere (ähnlich wie in der Vorlesung) über die Liste *soprano_note_names* und suche nach dem Fugenthema *fugue_subject*.\
Speichere für jedes Fugenthema den entsprechen Index, zu dem es gefunden wird, in einer Liste ab.

```python
subject_found_for_these_indices = []

number_steps = len(soprano_note_names) - len(fugue_subject) + 1

for i in range(number_steps):

    if fugue_subject == soprano_note_names[i : i+len(fugue_subject)]:
        
        #subject found!
        subject_found_for_these_indices.append(i)
```

```python
print(subject_found_for_these_indices)
```

9) Nutze die gefundenen Indizes, um aus der Noten-Liste *soprano_notes* die Offsets herauszulesen, zu denen ein Fugenthema gefunden wurde.

```python
subject_found_for_these_offsets = []

for index in subject_found_for_these_indices:

    first_note_of_subject = soprano_notes[index]
    offset_of_first_note_of_subject = first_note_of_subject.offset  

    subject_found_for_these_offsets.append(offset_of_first_note_of_subject)
```

```python
print(subject_found_for_these_offsets)
```

10) Wissend, dass die Fuge im 4/4-Takt steht, schließe aus den Offsets auf die Taktzahl.

```python
subject_found_for_these_measures = []

for offset in subject_found_for_these_offsets:
    measure_number = offset/4 + 1
    subject_found_for_these_measures.append(measure_number)
```

```python
print(subject_found_for_these_measures)
```

11) Wie könnte man das Fugenthema auch in den restlichen Stimmen der Fuge finden?

- Über *bach_score.parts* iterieren
- Den Code der Fragen 4 bis 9 "wiederverwenden" (z. B. durch Nutzen des Codes in Form von Funktionen)

12. Alternativ hätte man – statt der Offsets – auch gleich die Taktzahlen aus den *Note*-Objekte herauslesen können.\
Dazu kann die *Note*-Methode *.measureNumber* genutzt werden.

```python
subject_found_for_these_measures_alternative = []

for index in subject_found_for_these_indices:
    first_note_of_subject = soprano_notes[index]
    measure_number_of_first_note_of_subject = first_note_of_subject.measureNumber  

    subject_found_for_these_measures_alternative.append(measure_number_of_first_note_of_subject)
```

```python
print(subject_found_for_these_measures_alternative)
```

### Methode 2: Note-Extraction mithilfe von *.recurse()*

Die Methode *.recurse()*. ist eine weitere wichtige Methode, um auf in einem Stream enthaltene musikalische Elemente und Parameter zugreifen zu können. \
*.recurse()* erlaubt es, über alle Elemente eines Streams und dessen Substreams zu iterieren. \
Dabei geht – anders als bei der Methode *flatten()* – die verschachtelte Struktur von Streams und Substreams nicht verloren.\
Es bleiben also insbesondere alle Takte enthalten!

12) Nutze dieses Mal die *recurse()*-Methode – zusammen mit der Methode *getElementsByClass(...)* – um alle m21.note.Note-Objekte aus der Sopran-Stimme zu extrahieren.

```python
soprano_notes2 = list(soprano_part.recurse().getElementsByClass(m21.note.Note))
```

```python
print(soprano_notes2)
```

13) Warum funktioniert die Offset-Methode aus den Fragen 9 und 10 jetzt nicht mehr?

- *recurse()* bewahrt alle Streams und Substreams.
- Offsets aller Objekte werden in *music21* immer in Bezug auf den Beginn des nächst-übergeordneten Streams, in dem sie sich befinden angegeben: für Noten sind das Takte (*Measure*-Strams)
