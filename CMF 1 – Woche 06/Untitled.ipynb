```python
from minineedle import needle, smith, core
```

```python
sequence1 = "some trail snacks"
sequence2 = "a snail track"
```

```python
alignementNW = needle.NeedlemanWunsch(sequence1, sequence2)
alignementNW.change_matrix(core.ScoreMatrix(match = 1, miss = -1, gap = -2))
alignementNW.align()
print(alignementNW, alignementNW.get_score())
```
