### Motiv Suche 1

```python
clara_score = m21.corpus.parse("schumann_clara/opus17/movement3.xml")
```

```python
for part in clara_score.parts:
    print(part.partName)
```

```python
motiv = ["B","E","E","E","D#"]
clara_cello = list(clara_score.parts[1].flatten().getElementsByClass(m21.note.Note))
cello_notes = []
for elem in clara_cello:
    cello_notes.append(elem.name)
cello_notes
```

```python
for i in range(len(cello_notes)):
    if motiv == cello_notes[i : i+len(motiv)]:
        print(f"found at note {i}: \n{cello_notes[i : i+len(motiv)]}")
```
