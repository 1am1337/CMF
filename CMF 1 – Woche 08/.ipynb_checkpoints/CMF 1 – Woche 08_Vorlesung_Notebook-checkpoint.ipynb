### Dynamic Time Warping

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
sequence1 = "ABCDEF"
sequence2 = "ABCCCDEEF"
```

1) Berechnung einer *Cost-Matrix*:

```python
def cost_matrix(sequence1, sequence2):

    len_sequence1 = len(sequence1) 
    len_sequence2 = len(sequence2)

    CM = np.zeros([len_sequence1, len_sequence2])

    for i in range(len_sequence1):
        for j in range(len_sequence2):

            if sequence1[i] == sequence2[j]:
                CM[i, j] = 0
            else:
                CM[i, j] = 1

    return CM
```

```python
CM = cost_matrix(sequence1, sequence2)
```

```python
plt.figure(figsize=(4, 3))
plt.imshow(CM, cmap='binary', origin='lower')
plt.show()
```

2) Bestimmung eines *Warping-Pfads* mithilfe von *librosa*:

```python
ACM, warping_path = librosa.sequence.dtw(C = CM,
                                         step_sizes_sigma = np.array([[1,0], [0,1], [1,1]]))
```

```python
plt.figure(figsize=(4, 3))
plt.imshow(ACM, cmap='binary', origin='lower')
plt.show()
```

```python
warping_path = np.array(warping_path) 

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(CM, cmap='gray_r', origin='lower', aspect='equal')
plt.plot(warping_path[:, 1], warping_path[:, 0], marker='o', color='r')
#plt.clim([0, np.max(CM)])
#plt.colorbar()
plt.title('Optimal warping path on CM')

plt.subplot(1, 2, 2)
plt.imshow(ACM, cmap='gray_r', origin='lower', aspect='equal')
plt.plot(warping_path[:, 1], warping_path[:, 0], marker='o', color='r')
#plt.clim([0, np.max(ACM)])
#plt.colorbar()
plt.title('Optimal warping path on ACM')

plt.tight_layout()
```
