### Aufgabe 03.

Die Python-Library *music21* erlaubt es, die gängigen symbolischen Musik-Formate (insbesondere *MIDI* und *MusicXML*) zu importieren, zu analysieren und zu bearbeiten.

Nutze *music21* in dieser Aufgabe, um eine MIDI- und eine MusicXML-Datei zu importieren und als Partitur anzuzeigen.

```python
import music21 as m21
```

```python
# Importieren der MusicXML-Datei "Aufgabe 03_MusicXML.musicxml" als "music21-Score" mithilfe der Methode ".converter.parse"
aufgabe03_musicxml = m21.converter.parse("Aufgabe 03_MusicXML.musicxml")

# Anzeige des "music21-Scores" "aufgabe03_musicxml" als Partitur mithilfe der Methode ".show()"
aufgabe03_musicxml.show()
```

```python
# Importieren der MIDI-Datei "Aufgabe 03_MusicXML.mid" als "music21-Score" mithilfe der Methode ".converter.parse"
aufgabe03_midi = m21.converter.parse("Aufgabe 03_MIDI.mid")

# Anzeige des "music21-Scores" "aufgabe03_midi" als Partitur mithilfe der Methode ".show()"
aufgabe03_midi.show()
```
