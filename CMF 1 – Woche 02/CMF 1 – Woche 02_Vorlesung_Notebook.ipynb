### Flattening und Recursion

```python
for thing in measure1:
    print(thing)
```

```python
for object in measure1:
    if type(object) == m21.note.Note:
        print(object.pitch)
```

```python
for object in measure1:
    if isinstance(object, m21.note.Note):
        print(object.pitch)
```

```python
for note in part1.recurse().getElementsByClass(m21.note.Note):
    print(note.pitch)
```

```python
for measure in part1.recurse().getElementsByClass(m21.stream.Measure):
    print(measure)
```

```python
part1.show('text')
```

```python
part1_flattened = part1.flatten()
```

```python
part1_flattened.show('text')
```

```python
for note in part1_flattened.getElementsByClass(m21.note.Note):
    print(note.offset)
```

Die Methode *.flatten()* ist eine wichtige Methode, um auf alle in einem Stream enthaltenen musikalischen Elemente und Parameter zugreifen zu können. \
*.flatten()* erlaubt es, einen *Score* "abzuflachen", in dem es die verschachtelte Strukturen von Streams und Substreams in einen einzigen Stream umwandelt. \
Der resultierende Stream enthält dann nur noch *music21*-Objekte, die keine Stream-Subklassen sind.

Die Methode *.recurse()*. ist eine weitere wichtige Methode, um auf in einem Stream enthaltene musikalische Elemente und Parameter zugreifen zu können. \
*.recurse()* erlaubt es, über alle Elemente eines Streams und dessen Substreams zu iterieren. \
Dabei geht – anders als bei der Methode *flatten()* – die verschachtelte Struktur von Stream und Substream nicht verloren.

### Motiv-Suche I

```python
some_piece = m21.corpus.parse("schumann_clara/opus17/movement3.xml")
```

```python
for part in some_piece.parts:
    print(part.partName)
```

```python
some_piece.parts[1].partName
```

```python
motif = ['B', 'E', 'E', 'E', 'D#'] # E- wäre Es
```

```python
flattened_cello = some_piece.parts[1].flatten()
```

```python
cello_notes = list(flattened_cello.getElementsByClass(m21.note.Note))
```

```python
cello_note_names = []
for note in cello_notes:
    cello_note_names.append(note.name)
```

```python
for i in range(len(cello_note_names)):
    if motif == cello_note_names[i : i+len(motif)]:
        print(cello_note_names[i : i+len(motif)])
        print("found")
```
