### Frage: Wie viele Moll-Werke im Bach-Korpus enden mit einer Picardischen Terz?

```python
import music21 as m21
```

```python
# Laden aller MusicXML-Dateien des music21-Bach-Korpus
bach_corpus = m21.corpus.search('bach', fileExtensions='xml')
```

Auflistung aller Moll-Werke im Bach-Korpus

```python
# Liste für alle Stücke, die in Moll stehen
scores_of_pieces_in_minor = []

for piece in bach_corpus:

    # Erstellen eines m21-Scores für das aktuelle Stück
    piece_score = piece.parse()

    # Steht das aktuelle Stück in Moll?
    if piece_score.analyze('key').mode == 'minor':
        # Wenn ja, dann füge das Stück zur Liste hinzu
        scores_of_pieces_in_minor.append(piece_score)
```

```python
print("Der Bach-Korpus enthält %i Stücke." % len(bach_corpus))
print("Der Bach-Korpus enthält %i Stücke in Moll." % len(scores_of_pieces_in_minor))
```

Bestimmung der Stücke, die mit einer Picardischen Terz enden

```python
# Liste für alle Stücke, die mit einer Picardischen Terz enden
pieces_with_picardy_third = []

for score in scores_of_pieces_in_minor:

    current_score_chordified = score.chordify()
    
    # Herauslesen aller Chord-Objekte aus dem chordifizierten Score
    current_score_chords = current_score_chordified.recurse().getElementsByClass(m21.chord.Chord)

    # Analyse des letzten Akkords
    if current_score_chords[-1].commonName == 'major triad':
        
        pieces_with_picardy_third.append(score)
```

```python
print("%i Stücke enden mit einer Picardischen Terz." % len(pieces_with_picardy_third))
```

### Frage: Ist die Angabe des jeweils häufigsten Akkords ein geeignetes Werkzeug, um die Tonart eines Stücks (z. B. von J. S. Bach) zu bestimmen?

```python
import music21 as m21
from collections import Counter
```

1) Lade alle MusicXML-Dateien des music21-Bach-Korpus.

```python
bach_corpus = m21.corpus.search('bach', fileExtensions='xml')
```

2) Teste, welche Ergebnisse die m21.chord.Chord-Methoden *.root()*, *.root().name* und *.quality* liefern.

```python
my_chord = m21.chord.Chord(["E3", "G3", "C4"])
print(my_chord.root())
print(my_chord.root().name)
print(my_chord.quality)
```

3) Schreibe eine Funktion *chord_counter*, die alle unterschiedlichen Akkorde, die in einem Stück vorkommen, zählt und auflistet. \
Nutze dazu die Funktion *Counter* aus der Python-Library *collections* und die *Counter*-Methode *.items()*

```python
def chord_counter(piece):

    # Erstelle ein m21-Score des Stücks
    score = piece.parse()

    # Reduziere den Score zu einem Akkord-Stream mithilfe der Methode chordify
    score_chordified = score.chordify()

    # Extrahiere alle Akkorde mit Methode ".getElementsByClass"
    score_chords = score_chordified.recurse().getElementsByClass(m21.chord.Chord)

    # Liste für alle im Stück vorkommende unterschiedliche Akkordnamen
    all_chord_names = []

    for chord in score_chords:
        all_chord_names.append(chord.root().name + " " + chord.quality)

    # Häufigkeit der verschiedenen Akkordnamen
    counted_chord_names = Counter(all_chord_names)
    list_counted_chord_names = list(counted_chord_names.items())

    return list_counted_chord_names
```

4) Teste die Funktion an einem beliebigen Stück aus dem Bach-Korpus.

```python
chord_counter(bach_corpus[0])
```

5) Schreibe eine Funktion *sorted_chord_counter*, die den Output der Funktion *chord_counter* nach Häufigkeit der Akkorden sortiert.

```python
def sorted_chord_counter(piece):

    list_counted_chord_names = chord_counter(piece)
    
    sorted_list_counted_chord_names = sorted(list_counted_chord_names, key = lambda item: item[1], reverse=True)

    return sorted_list_counted_chord_names
```

6) Teste die Funktion an einem beliebigen Stück aus dem Bach-Korpus.

```python
sorted_chord_counter(bach_corpus[0])
```

7) Nutze diese Funktion und schreibe eine weitere Funktion *most_frequent_chord*, die den häufigsten Akkord eines Stücks ausgibt.

```python
def most_frequent_chord(piece):

    sorted_list_counted_chord_names = sorted_chord_counter(piece)
    
    most_frequent_chord_name = sorted_list_counted_chord_names[0][0]

    return most_frequent_chord_name
```

8) Teste die Funktion an einem beliebigen Stück aus dem Bach-Korpus.

```python
most_frequent_chord(bach_corpus[0])
```

9) Bestimme den jeweils häufigsten Akkord für die 20 ersten Stücke im Bach-Korpus. Speichere alle gefundenen Akkorde in einer Liste *most_frequent_chords* ab.

```python
most_frequent_chords = []

for piece in bach_corpus[:20]:

    most_frequent_chords.append(most_frequent_chord(piece))
```

```python
print(most_frequent_chords)
```

11. Zähle die Übereinstimmungen zwischen der Tonart der 20 ersten Stücke des Bach-Korpus und dem jeweils häufigsten Akkord. \
Hinweis: In music21 gibt es unterschiedliche Konvention für die Groß- und Kleinschreibung bei Tonarten und Akkordnamen. Deshalb ist es für unsere Anwendung sinnvoll, alle Strings in Uppercase-Strings umzuwandeln, um die Strings vergleichen zu können.

```python
# Counter, der die Übereinstimmungen zwischen der Tonart eines Stücks und dem häufigsten Akkord des Stücks zählt
key_chord_match = 0

# Iteriere über die 20 ersten Stücke im Bach-Korpus
for i, piece in enumerate(bach_corpus[:20]):

    if most_frequent_chords[i].upper() == str(piece.parse().analyze('key')).upper():
        
        key_chord_match += 1
```

```python
print("Bei %i Stücken gibt es eine Übereinstimmung zwischen der Tonart des Stücks und dem häufigsten Akkord des Stücks." % key_chord_match)
```

12) Bestimme, bei viel Prozent der Stücken es eine Übereinstimmung zwischen der Tonart des Stücks und dem häufigsten Akkord des Stücks gibt.

```python
key_chord_match_percent = (key_chord_match/20)*100
```

```python
print("Bei %0.1f Prozent der Stücken in Moll gibt es eine Übereinstimmung zwischen der Tonart des Stücks und dem häufigsten Akkord des Stücks." % key_chord_match_percent)
```
