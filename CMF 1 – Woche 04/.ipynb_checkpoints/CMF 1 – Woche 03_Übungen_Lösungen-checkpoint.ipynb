### Übungsaufgabe 01. Fugenthemen als rhythmische Patterns.

In der Vorlesung haben wir die Chor-Fuge *Omnes Generationes* von J. S. Bach analysiert, indem wir das Fugenthema als Intervall-Pattern dargestellt haben. \
Das Ziel dieser Übungsaufgabe ist es zu überprüfen, ob es ebenfalls möglich ist, alle Themen-Einsätze zu erkennen, indem man das Fugenthema als ryhthmisches Pattern betrachtet – d. h. ohne Berücksichtigung der Tonhöhen.

```python
import music21 as m21
import matplotlib.pyplot as plt
```

1) Extrahiere – wie in der Vorlesung – alle in der ersten Sopran-Stimme vorkommenden Noten.

```python
omnes = m21.converter.parse('bach_omnes-generationes.mxl')
soprano_part = omnes.parts[0]
```

```python
# Erstelle eine "flatted"-Version dieses Part-Stream
# (wichtig, um auf die Offsets aller Noten mit Bezug auf den Anfang des Stückes zugreifen zu können)
flattened_soprano_part = soprano_part.flatten()
# Extrahiere alle in der Sopran-Stimme enthaltenen Noten
soprano_notes = flattened_soprano_part.getElementsByClass(m21.note.Note)
```

2) Beschreibe die Funktion der *note.duration*-Methoden *type* und *quarterLength*.

```python
print(soprano_notes[0].duration.type)
print(soprano_notes[0].duration.quarterLength)
```

3) Stelle das Fugenthema und die Sopran-Stimme als eine Abfolge von rhythmischen Werten dar – durch "Vergessen" der Tonhöhen.\
*Hinweis:* Da es am Ende eines Themeneinsatzes oft zu rhythmischen Mutationen kommen kann, kann es sinnvoll sein, die letzte Note des Themas nicht zu berücksichtigen.

```python
soprano_rhythms = []

for note in soprano_notes:
    note_rhythm = note.duration.quarterLength
    soprano_rhythms.append(note_rhythm)
```

```python
subject_rhythms = [0.5, 0.5, 0.5, 0.5, 0.5, 0.25, 0.25, 0.5, 0.5]
```

4) Gleiche die Rhythmen des Fugenthemas mit den Rhythmen der Sopranstimme ab und versuche allein auf Grundlage des Rhythmus alle Themen-Einsätze zu erkennen.\
Vergleiche die gefundenen Themen-Einsätze mit den Ergebnissen aus der Vorlesung.\
*Hinweis:* Nutze dazu den Code aus der Vorlesung.

```python
def pattern_finder(rhythm_list, rhythmical_pattern):

    # Zu welchen Indezes kommt das Pattern rhythmical_pattern in der Liste rhythm_list vor:
    list_with_pattern_indices = []
    
    len_rhythm_list = len(rhythm_list)
    len_rhythmical_pattern = len(rhythmical_pattern)

    for i in range(len_rhythm_list):

        if rhythm_list[i : i+len_rhythmical_pattern] == rhythmical_pattern:   
        
            list_with_pattern_indices.append(i)
            
    return list_with_pattern_indices
```

```python
subject_indices = pattern_finder(soprano_rhythms, subject_rhythms)
```

```python
print(subject_indices)
```

```python
subject_offsets = []

for index in subject_indices:
    subject_offsets.append(soprano_notes[index].offset)
```

```python
subject_measures = [offset/4 + 1 for offset in subject_offsets]
print(subject_measures)
```

```python
x = subject_measures
y = ["Soprano 1"]*len(x)

plt.style.use('dark_background')

fig, ax = plt.subplots(figsize=(10, 2.5))

ax.set_frame_on(False)

ax.scatter(x, y, c='plum', marker='|')
ax.set_xlabel('Measures', fontsize=10)

plt.show()
```
